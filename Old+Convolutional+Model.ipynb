```python
# Load data
train_volume = io.imread('Data/train-volume.tif')
train_labels = io.imread('Data/train-labels.tif')
test_volume = io.imread('Data/test-volume.tif')
```

```python
# Normalize function
def normalize(x):
    x = x.astype(np.float32)
    x /=255
    return x
```

```python
# Normalize data
train_volume = normalize(train_volume)
train_labels = normalize(train_labels)
test_volume = normalize(test_volume)
```

```python
# Batch function

# The data is currently a multidimensional numpy array (ndarray) with 3 dimensions. It has the shape 30 * 512 * 512, which
# means that the data exists of 30 greyscale pictures with dimension 512 pixels on 512 pixels. We want to make 30 batches of 1 
# picture each. We'll also have to reshape the data to pass it to convolutional layers in tensorflow (4 dimensions required).

def get_batches(data):
    data = np.reshape(data, (30, 512, 512, 1))
    data = np.split(data, 30)
    return data
```

```python
# Batch data
volume_batches = get_batches(train_volume)
labels_batches = get_batches(train_labels)
test_batches = get_batches(test_volume)
```

```python
# Construct a convolutional autoencoder

# Create placeholders
inputs = tf.placeholder(dtype = tf.float32, shape=(None, 512, 512, 1), name='inputs')
targets = tf.placeholder(dtype = tf.float32, shape=(None, 512, 512, 1), name='targets')
learning_rate = 0.001
dropout = 0.5

### Encoder
conv1 = tf.layers.conv2d(inputs, 32, 5, 1, padding='same', activation=tf.nn.relu)
# Now 512x512x32
maxpool1 = tf.layers.max_pooling2d(conv1, 2, 2, padding='same')
# Now 256x256x32
dropout1 = tf.layers.dropout(maxpool1, rate=dropout)
conv2 = tf.layers.conv2d(dropout1, 64, 5, 1, padding='same', activation=tf.nn.relu)
# Now 256x256x64
maxpool2 = tf.layers.max_pooling2d(conv2, 2, 2, padding='same')
dropout2 = tf.layers.dropout(maxpool2, rate=dropout)
# Now 128x128x64
conv3 = tf.layers.conv2d(dropout2, 128, 5, 1, padding='same', activation=tf.nn.relu)
# Now 128x128x128

encoded = tf.layers.max_pooling2d(conv3, 2, 2, padding='same')
# Now 64x64x128
dropout3 = tf.layers.dropout(encoded, rate=dropout)
    
### Decoder
upsample1 = tf.image.resize_nearest_neighbor(dropout3, (64,64))
# Now 64x64x128
conv5 = tf.layers.conv2d(upsample1, 128, 5, 1, padding='same', activation=tf.nn.relu)
# Now 64x64x128
dropout4 = tf.layers.dropout(conv5, rate=dropout)
upsample2 = tf.image.resize_nearest_neighbor(dropout4, (128,128))
# Now 128x128x128
conv6 = tf.layers.conv2d(upsample2, 64, 5, 1, padding='same', activation=tf.nn.relu)
# Now 128x128x64
dropout5 = tf.layers.dropout(conv6, rate=dropout)
upsample3 = tf.image.resize_nearest_neighbor(dropout5, (256,256))
# Now 256x256x64
conv7 = tf.layers.conv2d(upsample3, 32, 5, 1, padding='same', activation=tf.nn.relu)
# Now 256x256x32
dropout6 = tf.layers.dropout(conv7, rate=dropout)
upsample4 = tf.image.resize_nearest_neighbor(dropout6, (512,512))
# Now 512x512x32
    
logits = tf.layers.conv2d(upsample4, 1, 5, 1, padding='same', activation=None)
# Now 512x512x1
dropout7 = tf.layers.dropout(logits, rate=dropout)
    
decoded = tf.sigmoid(dropout7, name='decoded')

# Loss and Optimizer
loss = tf.nn.sigmoid_cross_entropy_with_logits(labels=targets, logits=logits)
cost = tf.reduce_mean(loss)
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
## Function for training the Neural Network
def train_neural_network(session, optimizer, feature_batch, label_batch):
    session.run(optimizer, feed_dict = {inputs: feature_batch, targets: label_batch})

# Function for showing the stats
def print_stats(session, feature_batch, label_batch, cost):
    print("cost: ", session.run(cost, feed_dict = {inputs: feature_batch, targets: label_batch}))

# Hyperparameters
epochs = 50
batch_size = 1
          
### Train Model

# Create the session
sess = tf.Session()

# Initializing the variables
sess.run(tf.global_variables_initializer())

# Training cycle
for epoch in range(epochs):
    n_batches = len(volume_batches)
    for i in range(0, n_batches):
        train_neural_network(sess, optimizer, volume_batches[i], labels_batches[i])
        print('Epoch {:>2}, Batch {}:  '.format(epoch + 1, i), end='')
        print_stats(sess, volume_batches[i], labels_batches[i], cost) 
```
